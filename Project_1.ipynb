```python
import panel as pn
pn.extension('plotly')
import plotly.express as px
import pandas as pd
import hvplot.pandas
import matplotlib.pyplot as plt
import numpy as np
import os
from pathlib import Path
from dotenv import load_dotenv
```

```python
#stable coin data

busd_path = "Resources/busd.csv"
dai_path = "Resources/dai.csv"
usdc_path = "Resources/usdc.csv"
usdt_path = "Resources/usdt.csv"

busd_df = pd.read_csv(busd_path, infer_datetime_format=True)
dai_df = pd.read_csv(dai_path, infer_datetime_format=True)
usdc_df = pd.read_csv(usdc_path, infer_datetime_format=True)
usdt_df = pd.read_csv(usdt_path, infer_datetime_format=True)


busd_df.set_index('Date', inplace=True)
dai_df.set_index('Date', inplace=True)
usdc_df.set_index('Date', inplace=True)
usdt_df.set_index('Date', inplace=True)

# Drop the columns that are not need 

busd_df = busd_df.drop(columns=['Change %'])
dai_df = dai_df.drop(columns=['Change %'])
usdc_df = usdc_df.drop(columns=['Change %'])
usdt_df = usdt_df.drop(columns=['Change %'])


#join into one table

stable_coin_df = pd.concat([busd_df, dai_df, usdc_df, usdt_df], axis=1, join='outer', keys=('BUSD', 'DAI', 'USDC', 'USDT'))
stable_coin_df.dropna()
```
